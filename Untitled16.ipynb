```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("ECommerce Customer Behavior Analysis") \
    .getOrCreate()

print("Spark Session Created")
```

```python
df = spark.read.csv("ecommers-data.csv", header=True, inferSchema=True)
df.show(5)
```

```python
from pyspark.sql.functions import col

# Drop null values
df = df.dropna()

# Convert data types
df = df.withColumn("Quantity", col("Quantity").cast("int")) \
       .withColumn("UnitPrice", col("UnitPrice").cast("double"))

df.printSchema()
```

```python
from pyspark.sql.functions import expr

df = df.withColumn("TotalAmount", expr("Quantity * UnitPrice"))
df.show(5)
```

```python
from pyspark.sql.functions import expr

df = df.withColumn("TotalAmount", expr("Quantity * UnitPrice"))
df.show(5)
```

```python
top_customers = df.groupBy("CustomerID") \
    .sum("TotalAmount") \
    .withColumnRenamed("sum(TotalAmount)", "TotalSpent") \
    .orderBy(col("TotalSpent").desc())

top_customers.show(10)
```

```python
from pyspark.sql.functions import col

popular_products = df.groupBy("ProductNo") \
    .sum("Quantity") \
    .withColumnRenamed("sum(Quantity)", "TotalQuantity") \
    .orderBy(col("TotalQuantity").desc())

popular_products.show(10)
```

```python
from pyspark.sql.functions import lit

df = df.withColumn("Country", lit("India"))
```

```python
country_sales = df.groupBy("Country") \
    .sum("TotalAmount") \
    .withColumnRenamed("sum(TotalAmount)", "TotalSales") \
    .orderBy(col("TotalSales").desc())

country_sales.show()
```

```python
from pyspark.sql.functions import sum

customer_df = df.groupBy("CustomerID").agg(
    sum("Quantity").alias("TotalQuantity"),
    sum("TotalAmount").alias("TotalSpent")
)

assembler = VectorAssembler(
    inputCols=["TotalQuantity", "TotalSpent"],
    outputCol="features"
)

data = assembler.transform(customer_df)

kmeans = KMeans(k=3, seed=1)
model = kmeans.fit(data)

predictions = model.transform(data)

predictions.show()
```

```python
from pyspark.ml.evaluation import ClusteringEvaluator

evaluator = ClusteringEvaluator()
silhouette = evaluator.evaluate(predictions)

print("Silhouette Score:", silhouette)
```
